## 4. Exploratory Data Analytics

### EDA Using Data Visualization

#### Univariate Analytics 

- for categorical features - barplot
- for numeric features = histogram

```python
df.columns
```

```python
cat_cols = ['Geography','Gender','HasCrCard','IsActiveMember','Exited']
num_cols = ['CreditScore','Age','Tenure','Balance','NumOfProducts','EstimatedSalary']
```

```python
for col in cat_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(df[col])
    plt.show()
    
```

Analytics - 
    - almost 50% of customers are from France and rest are from Spain and Germany
    - approximately 55% of customers are Male and rest are female
    - approx 50% of customers are not actively using their bank account
    - almost 70% of all customers have creditcard with them
    - approx 20% of all customers left the bank

```python
for col in num_cols:
    plt.figure(figsize=(12,4))
    plt.hist(df[col])
    plt.title(col)
    plt.savefig("Graph_univariate_"+col+".png")
    plt.show()
```

Analytics - 
    - Estimated Salary and tenure has uniform distribution
    - Most of customers have age between 25 to 50, there seems to be some outliers having age>80
    - there are two groups, one having balance = 0, other having balance b/w 75K to 200k
    - most of customer have creditscore between 450 to 850
    - most of customesr have 1 or 2 as count of products, very less have 3 or 4 products from bank

## Bivariate Analytics


1. Numeric v/s Categorical - distplot - probability distribution of data (normalized histogram)
2. Categorical v/s categorical - countplot

```python
# Numeric v/s categorical
# creditscore v/s Exited
plt.figure(figsize=(12,5))
sns.histplot(df["CreditScore"][df["Exited"]==0],color='lightgreen')
sns.histplot(df["CreditScore"][df["Exited"]==1],color='lightblue')
plt.legend(["0","1"])
plt.show()
```

Analytics -
     - both the distributions of CreditScore for Customers with Exited = 0 and Exited = 1 are almost overlapping eachother, thus at every value creditscore there is almost similar density of customers leaving and not leaving the bank, thus credit score has no information to say why customers left the bank.

```python
# Numeric v/s categorical
# creditscore v/s Exited
plt.figure(figsize=(12,5))
sns.histplot(df["CreditScore"][df["Exited"]==0])
sns.histplot(df["CreditScore"][df["Exited"]==1])
plt.legend(["0","1"])
plt.show()
```

```python
# Numeric v/s categorical
# Age v/s Exited
plt.figure(figsize=(12,5))
sns.histplot(df["Age"][df["Exited"]==0])
sns.histplot(df["Age"][df["Exited"]==1])
plt.legend(["0","1"])
plt.xlim([40,90])
plt.show()
```

Analytics - 
    - out of two distributions of creditscore, for age<40, distrobution of age of customers not leaving is higher compared to distribution of age of customers leaving the bank where as for age more than 40, distribution of age of customers leaving the bank is higher compared to not leaving. Thus most of old age customers left the bank.

```python
# categorical v/s categorical 
plt.figure(figsize=(6,3))
sns.countplot(df['Geography'])
plt.title("All customers")
plt.show()
plt.figure(figsize=(6,3))
sns.countplot(df['Geography'][df["Exited"]==1])
plt.title("Exited Customers")
plt.show()
```

```python
out = pd.crosstab(df['Geography'],df['Exited'],margins=True)
out
```

```python
out[1]/out['All']
```

```python
# categorical v/s categorical 
plt.figure(figsize=(6,3))
sns.countplot(df['Gender'])
plt.title("All customers")
plt.show()
plt.figure(figsize=(6,3))
sns.countplot(df['Gender'][df["Exited"]==1])
plt.title("Exited Customers")
plt.show()
```

```python
out = pd.crosstab(df['Gender'],df['Exited'],margins=True)
out[1]/out['All']
```

Analytics - 
    - Female customers have higher dropout rate compared to the male customers

```python
# categorical v/s categorical 
plt.figure(figsize=(6,3))
sns.countplot(df['HasCrCard'])
plt.title("All customers")
plt.show()
plt.figure(figsize=(6,3))
sns.countplot(df['HasCrCard'][df["Exited"]==1])
plt.title("Exited Customers")
plt.show()
```

```python
out = pd.crosstab(df['HasCrCard'],df['Exited'],margins=True)
out[1]/out['All']
```

### Multivariate Analysis

- numeric v/s numeric v/s categorical - scatterplot
- numeric v/s categorical v/s categorical - pointplot/swarmplot/boxplot

```python
# numeric v/s numeric v/s categorical 
plt.figure(figsize=(12,10))
sns.scatterplot(x="Age",y="CreditScore",hue="Exited",data=df)
plt.show()
```

```python
sns.catplot(x="Age",y="CreditScore",hue='Exited',col='Geography',data=df,kind='strip')
plt.show()
```

```python
# numeric v/s categorical v/s categorical
# Age v/s Geography v/s Exited - pointplot
# x = categorical, y = numeric, hue = numeric
plt.figure(figsize=(12,5))
sns.pointplot(x="Geography",y='Age',hue='Exited',data=df)
plt.show()
```

## EDA Using Analytics

#### correlation analysis

<img src ="correlation.PNG">

```python
cor = df.corr()
```

```python
#use heatmap to visualize correlation
plt.figure(figsize=(12,10))
sns.heatmap(cor,annot=True,cmap="coolwarm")
plt.show()
```

- if corr> +0.5 = very good correlation
- if corr is b/w +0.1 to +0.5 = good


- if corr <-0.5 = very good correlation
- if corr is b/w -0.1 to -0.5 = good


- if corr is b/w -0.1 to +0.1 = bad correlation
